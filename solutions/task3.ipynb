## 1. Marketing channel efficiency

### What the data shows

Every channel in this dataset converts at exactly the same rate at every step of the funnel. Same drop-offs, same conversion rates, across every channel. That kind of uniformity doesn't exist in real acquisition data. It's the clearest sign we're working with something synthetic.

Taking the structure of the data at face value: we don't have enough to make real channel decisions. No spend data, no LTV, no retention, no clicks or impressions. Without those, we can see that users moved through a funnel, but we can't say whether it was efficient or which channel deserves more budget. That's not a minor gap. That's most of the picture.

### The one real finding: KYC

KYC completion is where we lose the most users, and it's consistent across every channel. If there's one area the product team should prioritize, it's this step. A better KYC experience would improve full-funnel conversion across all channels at once. That's a product investment with the highest potential.

### Recommendation

Since every channel converts identically, cost per acquisition is the only variable left. Scale the cheapest channels while maintaining volume once we have spend data. And invest in building the measurement infrastructure we actually need: spend data, lifetime value, retention, and behavioral signals from inside the product. We can't make good decisions without them.

## 2. Paid social campaign: the metal premium experiment

### The design problem

There's a problem with how this experiment was set up, and it makes the results unreliable.

Users who saw the metal premium creative were routed to the standard product paywall. The ad was premium. The experience was standard. Those two need to match if we want to measure whether the creative worked. Without that alignment, we're not testing creative performance. We're testing what happens when you put a premium message in front of a standard product, which is a different question entirely.

There's also no control group. Without one, there's no way to separate the campaign's effect from everything else going on at the same time.

### What the experiment actually needs

Product and marketing need to be aligned on setup before running anything. Users who see the premium creative should reach a premium product experience, no cross-contamination.

The conversion event also needs to match the campaign goal. For a traffic campaign, signup might work as an early signal. For a conversion campaign, first transaction is the right measure. Signup tells us someone created an account. First transaction tells us if they find value. This difference is what we need to justify conversion campaign CPAs.

A holdout group matters here too. Considering the privacy restrictions since ios14, user-level holdouts are not possible for most setups. Geo-based holdouts do not give us user-level information, but they still give directional signal on incremental uplift. That's better than no baseline at all. 

### Recommendation

Creative performance can't be evaluated until the product routing is fixed. Once it is, redesign the experiment with a proper holdout and a conversion event that matches the campaign objective. What users do after signing up matters as much as the signup itself. Retention, transaction volume, and engagement inside the product are the signals that tell us whether a campaign actually worked. Upper-funnel metrics alone tell half the story.

## 3. Attribution model

I won't give a definitive answer here, because it depends on things we need to discuss together.

1. What's the conversion-tail / consideration window for N26 users?
2. What's the acquisition size for each market on a daily basis?
3. Does behavior change in each market in terms of marketing activities (such as spend-levels), and also the user-bases?

The last-click attribution probably undervalues the channels doing early-funnel work. A data-driven model, one that distributes conversion credit based on actual user paths, would give a more accurate picture of what's contributing to acquisition. That said, it requires enough conversion volume to produce reliable results. Also, it will create a workload for our data science team. Budgets and DS-workload need to be discussed upfront.

This is worth a proper conversation rather than a quick call at the end of a case study.
